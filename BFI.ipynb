```python
from typing import Union

Number = Union[int, float]


def calculate_bfi(
    perf_selected: Number,
    baseline_perf: Number,
    perf_optimal: Number = 1.0
) -> float:
    """
    Calculate Baseline Fitness Improvement (BFI).

    BFI measures the normalized predictive improvement of a
    selected feature subset relative to the baseline performance.

    Parameters
    ----------
    perf_selected : float
        Predictive performance obtained using the selected features.

    baseline_perf : float
        Predictive performance obtained using the baseline feature set.

    perf_optimal : float, default=1.0
        Maximum achievable performance.

    Returns
    -------
    float
        BFI value clipped to the interval [0, 1].
    """

    perf_selected = float(perf_selected)
    baseline_perf = float(baseline_perf)
    perf_optimal = float(perf_optimal)

    # Check that there is room for improvement
    if perf_optimal <= baseline_perf:
        raise ValueError(
            "perf_optimal must be greater than baseline_perf."
        )

    # Calculate normalized improvement
    bfi = (
        (perf_selected - baseline_perf)
        /
        (perf_optimal - baseline_perf)
    )

    # Restrict BFI to [0, 1]
    return float(max(0.0, min(1.0, bfi)))
```

```python
# Example of using BFI 
```

```python
# Baseline model performance
baseline_perf = 0.8437

# Performance after feature selection
perf_selected = 0.8858

# Maximum possible performance
perf_optimal = 1.0


bfi = calculate_bfi(
    perf_selected=perf_selected,
    baseline_perf=baseline_perf,
    perf_optimal=perf_optimal
)

print(f"BFI: {bfi:.4f}")
```

```python
fsa_results = {
    "Chi-square": 0.8858,
    "ANOVA": 0.8785,
    "IG": 0.8725,
    "mRMR": 0.8569,
    "PearsonCorr": 0.8714,
    "T-test": 0.8785,
    "RFE": 0.8762,
    "SFS": 0.8819,
    "SBS": 0.8853,
    "Boruta": 0.8764,
    "GA": 0.8537,
    "LASSO": 0.8646,
    "Ridge": 0.8860,
    "ElasticNet": 0.8821
}

baseline_perf = 0.8437
perf_optimal = 1.0


print("FSA             BFI")
print("----------------------")

for name, perf_selected in fsa_results.items():

    bfi = calculate_bfi(
        perf_selected=perf_selected,
        baseline_perf=baseline_perf,
        perf_optimal=perf_optimal
    )

    print(f"{name:15s} {bfi:.4f}")
```
